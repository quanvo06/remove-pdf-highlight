GỌI LỆNH NEW_CLIENT

```python
import pandas as pd
from chembl_webresource_client.new_client import new_client as nc
```

SEARCH TỪ KHOÁ

```python
target = nc.target
target_query = target.search('Acetylcholinesterase')
targets = pd.DataFrame.from_dict(target_query)
targets
```

CHỌN HOMOSAPIEN

```python
selected_target = targets.target_chembl_id[1]
selected_target
```

CÁC THỬ NGHIỆM TRÊN Dopamine D4

```python
activity = nc.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

```python
activity = nc.activity
res = activity.filter(target_chembl_id=selected_target).filter(standard_type="IC50")
```

```python

df = pd.DataFrame.from_dict(res)
df.head()
```

```python
df
```

LƯU LẠI Ở CSV file hoatchat_data.csv

```python
df.to_csv('hoatchat_data.csv', index=False)
```

LOẠI BỎ CÁC DỮ LIỆU THIẾU THÔNG TIN

```python
df2 = df[df.standard_value.notna()]
df2 = df2[df.canonical_smiles.notna()]
df2
```

ĐẾM SỐ LƯỢNG THẬT SỰ

```python
len(df2.canonical_smiles.unique())
```

BỎ LẶP LẠI

```python
df2_nr = df2.drop_duplicates(['canonical_smiles'])
df2_nr
```

CHỌN 3 CỘT CHÍNH

```python
selection = ['molecule_chembl_id','canonical_smiles','standard_value']
df3 = df2_nr[selection]
df3
```

LƯU

```python
df3.to_csv('hoatchat_data_2_daxuly.csv', index=False)
```

ĐỌC FILE

```python
df4 = pd.read_csv('hoatchat_data_2_daxuly.csv')
```

PHÂN LOẠI HOẠT CHẤT

```python
bioactivity_threshold = []
for i in df4.standard_value:
  if float(i) >= 10000:
    bioactivity_threshold.append("inactive")
  elif float(i) <= 1000:
    bioactivity_threshold.append("active")
  else:
    bioactivity_threshold.append("intermediate")
```

THÊM VÀO DF

```python

bioactivity_class = pd.Series(bioactivity_threshold, name='class')
df5 = pd.concat([df4, bioactivity_class], axis=1)
df5
```

LƯU DF5

```python
df5.to_csv('hoatchat_data_3_daxuly_dachuanhoa.csv', index=False)
```
